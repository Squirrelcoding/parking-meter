```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import h5py
from pathlib import Path

densities = []

pathlist = Path("../../data/car/car_heatmaps").glob("*")
for path in pathlist:
    with h5py.File(path, 'r') as f:
        matrix = f['density'][:]
        densities.append(matrix.sum())
```

```python
import pandas as pd

pd_densities = pd.Series(densities)
```

```python
pd_densities
```

```python
sns.set_theme()

sns.displot(pd_densities)
```
