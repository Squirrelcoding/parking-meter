```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("output.csv")
```

```python
fig, ax = plt.subplots()

ax.scatter(df['length'], df['width'])
```

```python
fig, ax = plt.subplots()

ax.hist(df['length'], bins=20)
```

```python
fig, ax = plt.subplots()

ax.hist(df['width'], bins=20)
```

```python
import numpy as np
import pylab
import scipy.stats as stats

stats.probplot(df['length'], dist="norm", plot=pylab)
pylab.show()
```

```python
import numpy as np
import pylab
import scipy.stats as stats

stats.probplot(df['width'], dist="norm", plot=pylab)
pylab.show()
```

```python
import statsmodels.api as sm
model = sm.OLS(df['width'], df['length'])
results = model.fit()

results.summary()
```

```python
mean = np.mean(df, axis=0)
cov = np.cov(df, rowvar=0)

mean, cov
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

y = multivariate_normal(mean=mean, cov=cov)
y.rvs()
```
